```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
data = pd.read_csv('/kaggle/input/recommendation-of-courses-for-students/train_M.csv')
data
```

```python
data_course_id = data['course_id'].value_counts()
data_course_id
```

```python
data_course_id.keys()
```

```python
submit = pd.read_csv('/kaggle/input/recommendation-of-courses-for-students/sample_submission_M.csv')
submit
```

```python
for ind, user_id in enumerate(tqdm(submit['user_id'])):
    dt_user_id = data[data['user_id'] == user_id]['course_id'].tolist()
    point = 0
    for el in data_course_id.keys():
        if el not in dt_user_id:
            point += 1
            submit.loc[ind, f"course_id_{point}"] = el
        if point == 3:
            break
    if point < 3:
        for i in range(3):
            submit.loc[ind, f"course_id_{i + 1}"] = data_course_id.keys()[i]
```

```python
submit.to_csv('recommendation-of-courses.csv', index=False)
```
